```python
# Step 1: Prompt the user
num_str = input("Enter a positive integer: ")

# Step 2: Validate input
if not num_str.isdigit():
    print("Invalid input. Please enter a positive integer.")
else:
    # Step 3: Initialize dictionary
    counts = {}

    # Step 4: Iterate over each character
    for ch in num_str:
        counts[ch] = counts.get(ch, 0) + 1

    # Step 5: Print each digit and its frequency in ascending order
    for digit in sorted(counts):
        print(f"Digit {digit}: {counts[digit]} time(s)")
```
