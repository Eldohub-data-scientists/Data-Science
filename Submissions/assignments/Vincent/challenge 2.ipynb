```python
import random

# Step 2: Generate random number
secret = random.randint(1, 50)

# Step 3: Initialize attempts
attempts = 0
max_attempts = 5

# Step 4: Loop for user guesses
while attempts < max_attempts:
    try:
        guess = int(input(f"Attempt {attempts + 1}/{max_attempts}: Your guess? "))
    except ValueError:
        print("Invalid input. Please enter an integer.")
        continue  # Skip incrementing attempts for invalid input

    attempts += 1

    if guess == secret:
        print(f"ðŸŽ‰ Correct! You guessed the number in {attempts} tries.")
        break
    elif guess < secret:
        print("Too low! Try again.")
    else:
        print("Too high! Try again.")
else:
    # Step 5: If user didn't guess correctly in time
    print(f"âŒ Out of attempts. The number was {secret}.")
```
